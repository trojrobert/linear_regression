```python
import matplotlib.pyplot as plt 
import matplotlib as mpl 
import pandas as pd 
import numpy as np 
import seaborn as sns 
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

##### Linear regression for data with 1 feature(independent variable)

```python
#import the data 
data = pd.read_csv('challenge_dataset.txt', names = ['features', 'target'])
#print(data)
```

```python
#plot the data 
sns.regplot(x='features', y='target', data = data, fit_reg=False )
plt.show()
```

```python
#divide the data into training and test data 
x_train, x_test, y_train, y_test = np.asarray(train_test_split(data['features'],data['target'],test_size=0.1))

#Create the model 
reg = LinearRegression()

#Fit the model by training it 
reg.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
```

```python
print('Score', reg.score(x_test.values.reshape(-1,1),y_test.values.reshape(-1,1)))
```

```python
# Plot regression 
x_line = np.arange(5,25).reshape(-1,1)
sns.regplot(x=data['features'], y=data['target'], data = data, fit_reg=False )
plt.plot(x_line,reg.predict(x_line))
plt.show()
```
